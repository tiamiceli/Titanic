```python
import numpy as np
#import matplotlib.pyplot as plt
import pandas as pd
#%matplotlib inline

from ipywidgets import interact
from IPython.display import display

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import FixedTicker
output_notebook()
```

```python
train = pd.read_csv("/Users/miceli/Data/Titanic/train.csv")
```

```python
print train.columns
```

```python
train.fillna(value=-10,inplace=True)
#mylist = train.Parch.tolist()
#b = type(mylist[1])==np.int64


#print mylist
train.Age
```

```python
hist_survived, edges_survived = np.histogram(train.Survived.tolist(),bins=2,range=(0,2))
print edges_survived
h_survived = figure(title="Number of Passengers that Died or Survived", x_axis_label="Died (0) or Survived (1)", y_axis_label="Number of Passengers", tools="save")
h_survived.quad(top=hist_survived, bottom=0, left=edges_survived[:-1], right=edges_survived[1:], fill_color="red", line_color=None)
h_survived.xaxis.ticker=FixedTicker(ticks=[0,1])
show(h_survived)
```

```python
hist_pclass, edges_pclass = np.histogram(train.Pclass.tolist(),bins=3, range=(1,4))
h_pclass = figure(title="Passenger's Ticket Class", x_axis_label="Passenger's Ticket Class", y_axis_label="Number of Passengers", tools="save")
h_pclass.quad(top=hist_pclass, bottom=0, left=edges_pclass[:-1], right=edges_pclass[1:], fill_color="orange", line_color=None)
h_pclass.xaxis.ticker=FixedTicker(ticks=[1,2,3])
show(h_pclass)
```

```python
hist_sibsp, edges_sibsp = np.histogram(train.SibSp.tolist(),bins=8,range=(0,8))
h_sibsp = figure(title="Number of Siblings or Spouses Aboard", x_axis_label="Number of Siblings or Spouses of Passenger", y_axis_label="Number of Passengers", tools="save")
h_sibsp.quad(top=hist_sibsp, bottom=0, left=edges_sibsp[:-1], right=edges_sibsp[1:], fill_color="gold", line_color=None)
h_sibsp.xaxis.ticker=FixedTicker(ticks=[0,1,2,3,4,5,6,7])
show(h_sibsp)
```

```python
hist_parch, edges_parch = np.histogram(train.Parch.tolist(),bins=7,range=(0,7))
h_parch = figure(title="Number of Parents or Children Aboard", x_axis_label="Number of Parents or Children of Passenger", y_axis_label="Number of Passengers", tools="save")
h_parch.quad(top=hist_parch, bottom=0, left=edges_parch[:-1], right=edges_parch[1:], fill_color="olivedrab", line_color=None)
h_parch.xaxis.ticker=FixedTicker(ticks=[0,1,2,3,4,5,6])
show(h_parch)
```

```python
hist_age, edges_age = np.histogram(train.Age.tolist(),bins=92,range=(-10,81))
h_age = figure(title="Age of Passenger (-10 = not available)", x_axis_label="Passenger's Age", y_axis_label="Number of Passengers", tools="save")
h_age.quad(top=hist_age, bottom=0, left=edges_age[:-1], right=edges_age[1:], fill_color="steelblue", line_color=None)
h_age.xaxis.ticker=FixedTicker(ticks=[-10,0,10,20,30,40,50,60,70,80])
show(h_age)
```

```python
hist_fare, edges_fare = np.histogram(train.Fare.tolist(),bins=104, range=(0,520))
h_fare = figure(title="Ticket Fare", x_axis_label="Ticket Fare", y_axis_label="Number of Passengers", tools="save")
h_fare.quad(top=hist_fare, bottom=0, left=edges_fare[:-1], right=edges_fare[1:], fill_color="darkviolet", line_color=None)
h_fare.xaxis.ticker=FixedTicker(ticks=[0,50,100,150,200,250,300,350,400,450,500,550,600,650])
show(h_fare)
```
