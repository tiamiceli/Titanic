```python
import numpy as np
#import matplotlib.pyplot as plt
import pandas as pd
#%matplotlib inline

from ipywidgets import interact
from IPython.display import display

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import FixedTicker
output_notebook()
```

```python
train = pd.read_csv("/Users/miceli/Data/Titanic/train.csv")
```

```python
print train.columns
```

```python
train.fillna(value=-1,inplace=True)
#mylist = train.Parch.tolist()
#b = type(mylist[1])==np.int64


#print mylist
type(train.Age[1])

```

```python
hist_survived, edges_survived = np.histogram(train.Survived.tolist(),bins=2)
h_survived = figure(title="Survived (0=died, 1=lived)", tools="save")
h_survived.quad(top=hist_survived, bottom=0, left=edges_survived[:-1], right=edges_survived[1:], fill_color="red", line_color="red")

#show(h_survived)
```

```python
hist_pclass, edges_pclass = np.histogram(train.Pclass.tolist(),bins=3)
h_pclass = figure(title="Passenger Class", tools="save")
h_pclass.quad(top=hist_pclass, bottom=0, left=edges_pclass[:-1], right=edges_pclass[1:], fill_color="orange", line_color="orange")

#show(h_pclass)
```

```python
myHistCollection = []
for col,col_name in enumerate(train.columns):
    mylist = np.array(train[col_name].tolist())
    if type(mylist[1])==np.int64:
        print "int64: "+col_name
    if type(mylist[1])==np.float64:
        print "float64: "+col_name
    if type(mylist[1])==str:
        print "str: "+col_name
    
    #make histograms of int variables
    if (type(mylist[1])==np.int64):
        print "   min = "+str(np.amin(mylist))+ " max = "+str(np.amax(mylist))
        minrange = np.amin(mylist)
        maxrange = np.amax(mylist)+1
        numbins = maxrange-minrange 
        hist_col_name, edges_col_name = np.histogram(mylist,bins=numbins,range=(minrange,maxrange))
        h_ = figure(title=col_name, tools="save")
        h_.quad(top=hist_col_name, bottom=0, left=edges_col_name[:-1], right=edges_col_name[1:])
        #h_.xaxis.minor_tick_line_color = None
        h_.xaxis.ticker=FixedTicker(ticks=np.linspace(minrange,maxrange-1,numbins))
        myHistCollection.append(h_)
        
```

```python
for h in myHistCollection:
    show(h)
```
